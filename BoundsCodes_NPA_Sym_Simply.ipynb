```python
import numpy as np
import math
import itertools
import picos as pic
```

```python
def SDP(n,K,d,p,D):

    ##### SDP #####

    T=2**p

    pic.available_solvers()

    Ap = pic.RealVariable("Ap", (n+1,n+1))
    
    sdp = pic.Problem()
   
    sdp.set_objective('find') 


    sdp.add_constraint(Ap[0,0]==K**4)
    sdp.add_constraint(Ap[0,n]==K**3)
    
    
    for i in range(d):
        sdp.add_constraint((K*Ap[0,n-i]-Ap[0,i])== 0)
    

    ### A_j>=0 ###
    
    for m in range(n+1):
        A=0
        for j in range(m+1):
            A+=(-1)**(m-j)*D**(j)*math.comb(n-j,n-m)*Ap[0,j]
        sdp.add_constraint(A>=0)            
    print(sdp)
    sol = sdp.solve(solver='mosek', primals=True,verbosity=True)
    
    print(np.round(np.array(Ap),2))
    
        
    if sol.problemStatus == "feasible":
        return True                    
    elif sol.problemStatus == "infeasible":
        return False
    else:
        return sol.problemStatus 

    
D=2

Codes=[
[1,0,1],[1,1,1],
[2,0,2],[2,1,1],[2,2,1],
[3,0,2],[3,1,1],[3,2,1],[3,3,1],
[4,0,2],[4,1,2],[4,2,2],[4,3,1],[4,4,1],
[5,0,3],[5,1,3],[5,2,2],[5,3,1],[5,4,1],[5,5,1],
[6,0,4],[6,1,3],[6,2,2],[6,3,2],[6,4,2],[6,5,1],[6,6,1],
[7,0,3],[7,1,3],[7,2,2],[7,3,2],[7,4,2],[7,5,1],[7,6,1],[7,7,1],
]
Codes=[[6,2,2]]
Codes=[[7,7,1]]
#Codes=[[6,6,1]]
Codes=[[10, 8, 2]]
#### Existing qubit codes ####

for code in Codes:
    
    n,K,d=code
    p=n
    check=SDP(n,2**K,d,p,D)
    
    if check != True:
        print("%s=%s" % (code,check))

print("Done")



#### Not existing qubit codes (distance+1) ####

for code in Codes:
    
    n,K,d=code
    p=n
    check=SDP(n,2**K,d+1,p,D)
    
    if check != False:
        print("[%s,%s,%s]=%s" % (code[0],code[1],code[2]+1,check))

print("Done")
```
